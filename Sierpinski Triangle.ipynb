```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vertices = np.array([[1,1], [3,1], [2,3]])
plt.scatter(vertices[:,0], vertices[:,1])
plt.show()
```

```python
random_point = np.array([1.4, 1.2])
n_points = 50000
points_cache = np.zeros((n_points+4, 2))

points_cache[0:3] = vertices
points_cache[3] = random_point
for i in range(0, n_points):
    x = np.random.randint(3)
    random_point = 0.5*(np.add(random_point, vertices[x]))
    points_cache[i+4] = random_point
    
```

```python
plt.figure(figsize=(8,6))
plt.scatter(points_cache[:,0], points_cache[:,1], s=0.05)
plt.show()
```
